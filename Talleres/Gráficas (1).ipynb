```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color="magenta", linewidth=5,alpha=0.9, marker="o", markersize=20,markeredgewidth=5,markeredgecolor="lime",markerfacecolor="lightseagreen")
plt.xlabel("$x(cmÂ²)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1,)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y,color="darkviolet")
plt.plot(x,y2,color="k", dashes=[8,2,2,2,1,1])
plt.title("FunciÃ³n sen(4x)0-$2\pi$")
plt.grid(True)
plt.xlabel("Tiempo(seg)")
plt.ylabel("Amplitud de la onda")
plt.show()
```
