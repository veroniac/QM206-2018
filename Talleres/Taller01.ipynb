```python
n=30
a=1
b=1
Bibonacci=[]
#print(a)
#print(b) 
Bibonacci.append(a) 
Bibonacci.append(b) 
for i in range(n):
    c=a+b
    #print(c)
    Bibonacci.append(c)
    a=b
    b=c
print(Bibonacci) 
```
