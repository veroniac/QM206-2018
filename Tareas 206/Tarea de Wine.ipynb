## Base de Datos
### ¿Qué es wine datasets?
>El  conjunto  de  datos Wine original del repositorio de aprendizaje automático UCI  es un conjunto de datos de clasificación multiclase que tiene 13 atributos y 3 clases. Estos datos son el resultado de un análisis químico de vinos cultivados en la misma región en Italia pero derivados de tres cultivares diferentes. El análisis determinó las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos. Las clases 2 y 3 se utilizan como inliers y la clase 1 se reduce a 10 instancias para usar como ouliers.

El wine conjunto de datos contiene los resultados de un análisis químico de los vinos cultivados en un área específica de Italia. Tres tipos de vino están representados en las 178 muestras, con los resultados de 13 análisis químicos registrados para cada muestra. La variable se ha transformado en una variable categórica.

Los datos no contienen valores perdidos y constituyen solo datos numéricos, con una variable de destino de tres clases ( Type) para la clasificación.

```python
wine = datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_dataframe['target']=wine.target
x=wine_dataframe['total_phenols']
print(x.mean())
y=wine_dataframe['proline']
col=(wine_dataframe['target']+1)*2
print(y.mean())
total_phenol=wine_dataframe['total_phenols']*10
proline=wine_dataframe['proline']*10
```

```python
plt.scatter(x,y,c=col,s=60,alpha=0.5)
plt.xlabel('Fenoles totales (%)')
plt.ylabel('Prolina (%)')
plt.title('Porcentaje de Prolina vs Porcentaje de Fenoles')
plt.grid(True)
plt.show()
```

#### DESCRIPCIÓN DEL PRIMER GRÁFICO ####

En esta gráfica se observar el Porcentaje de Prolina vs Porcentaje de Fenoles los fenoles al igual que las prolinas son muy importantes para la fabricación del vino.

```python
wine = datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_dataframe['target']=wine.target
x=wine_dataframe['alcohol']
print(x.mean())
y=wine_dataframe['alcalinity_of_ash']
col=(wine_dataframe['target']+1)*2
print(y.mean())
ALCOHOL=wine_dataframe['alcohol']*10
ALCALINITY=wine_dataframe['alcalinity_of_ash']*10
```

```python
plt.scatter(x,y,c=col,s=70,alpha=0.6)
plt.xlabel('Alcohol (%)')
plt.ylabel('Alcalinidad de cenizas (%)')
plt.title('Alcalinidad vs Porcentaje de alcohol')
plt.grid(True)
plt.show()
```
