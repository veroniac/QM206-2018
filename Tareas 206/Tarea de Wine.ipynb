Estos datos son el resultado de un anÃ¡lisis quÃ­mico de vinos cultivados en la misma regiÃ³n en Italia pero derivados de tres cultivares diferentes. El anÃ¡lisis determinÃ³ las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos. Las clases 2 y 3 se utilizan como inliers y la clase 1 se reduce a 10 instancias para usar como ouliers.

El wine conjunto de datos contiene los resultados de un anÃ¡lisis quÃ­mico de los vinos cultivados en un Ã¡rea especÃ­fica de Italia. Tres tipos de vino estÃ¡n representados en las 178 muestras, con los resultados de 13 anÃ¡lisis quÃ­micos registrados para cada muestra. La variable se ha transformado en una variable categÃ³rica.

Los datos no contienen valores perdidos y constituyen solo datos numÃ©ricos, con una variable de destino de tres clases ( Type) para la clasificaciÃ³n.

```python
wine = datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data, columns=wine.feature_names)
wine_dataframe['target']=wine.target
x=wine_dataframe['total_phenols']
print(x.mean())
y=wine_dataframe['proline']
col=(wine_dataframe['target']+1)*2
print(y.mean())
total_phenol=wine_dataframe['total_phenols']*10
proline=wine_dataframe['proline']*10
```

```python
plt.scatter(x,y,c=col,s=60,alpha=0.5)
plt.xlabel('Fenoles totales (%)')
plt.ylabel('Prolina (%)')
plt.title('Porcentaje de Prolina vs Porcentaje de Fenoles')
plt.grid(True)
plt.show()
```
