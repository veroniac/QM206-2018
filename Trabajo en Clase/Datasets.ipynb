```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["A","B","C","D"])   #para nombrar las columnas se usa columns=
iris_dataframe["target"]=iris.target
iris_dataframe["A"]     #para imprimir solo la columna "A"
```

```python
iris=datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])   
iris_dataframe["target"]=iris.target
x=iris_dataframe["SL"]  #para sacar el promedio de los numeros en X
print(x.mean())
y=iris_dataframe["SW"]  #para sacar el promedio de los numero en y
col=(iris_dataframe["target"]+1)*2
print(y.mean())
PL=iris_dataframe["PL"]*50
PW=iris_dataframe["PW"]*150
```

```python
plt.scatter(x,y,c=col,s=PL,alpha=0.5)
plt.xlabel("Largo del sépalo(cm)")
plt.ylabel("Ancho del sépalo(cm)")
plt.title("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,c=col,s=PW,alpha=0.5)
plt.xlabel("Largo del sépalo(cm)")
plt.ylabel("Ancho del sépalo(cm)")
plt.title("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```
