```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class RegLin:
    def __init__(self):
        self.pendiente=0 #porpiedad no necesita parentesis
        self.intercepto=0
    def regresion_lineal(self,x,y): #metodo necesita parentesis
        '''Calcula la pendiente y el intercepto de la regresi√≥n lineal'''
        n=len(x)
        detM=n*sum(x*x)-sum(x)*sum(x)
        detA=sum(y)*sum(x*x)-sum(x*y)*sum(x)
        detB=n*sum(x*y)-sum(x)*sum(y)
        a=detA/detM
        b=detB/detM
        self.pendiente=b #valor asignado a cada propiedad
        self.intercepto=a
        return [a,b]
    def graficar(self,x,y):
        ycal=x*self.pendiente+self.intercepto
        plt.plot(x,y,'ro')
        plt.plot(x,ycal,'b--')
        plt.xlabel('t(s)')
        plt.ylabel('distancia(m)')
        plt.title('Distancia vs tiempo')
        plt.show()
```

```python
def regresion_lineal(x,y):
    n=len(x)
    detM=n*sum(x*x)-sum(x)*sum(x)
    detA=sum(y)*sum(x*x)-sum(x*y)*sum(x)
    detB=n*sum(x*y)-sum(x)*sum(y)
    a=detA/detM
    b=detB/detM
    return [a,b]
```

```python
x=np.array([0,1,2,3,4,5])
y=x*5+4+np.random.randn(6) 
reg=regresion_lineal(x,y)
ycal=x*reg[1]+reg[0]
ycal
```
