```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color='navy',linewidth=1,alpha=0.5,marker='o',markersize=6,markeredgecolor='black',markerfacecolor='salmon',markeredgewidth=2)
plt.xlabel('$x(cm²)$')
plt.ylabel('$Intensidad(\mu)$')
plt.title('Experimento No.1')
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y,color='darkcyan')
plt.plot(x,y2,color='b',dashes=[1,1,2,3])
plt.title('Función sen (4x)')
plt.xlabel('t (s)')
plt.ylabel('Amplitud de la onda')
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
r=np.random.rand(100)*50+50
plt.scatter(x,y,s=r, color='gray')
plt.show()
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=45, facecolor='coral', edgecolor='navy',density=True)
plt.title('Distribución de la Población')
plt.ylabel('Densidad (Hab/$m²)$')
plt.text(-3,0.35, r'$\mu=100,\ \sigma=15$')
plt.show()
```
